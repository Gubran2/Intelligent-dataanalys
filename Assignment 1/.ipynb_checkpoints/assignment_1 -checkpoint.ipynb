## Q6. Rank the overall temperature per month, sort from the coldest to thewarmest. List the top ten (10) coldest months in the dataset. Format theoutput as: YEAR/Month: temp, e.g.“2010/January: -5.0”

```python
df['Year'] = df['Datum'].dt.year
df['Month'] = df['Datum'].dt.month
monthly_avg_temp = df.groupby(['Year', 'Month'])['Lufttemperatur'].mean().reset_index()
coldest_months = monthly_avg_temp.nsmallest(10, 'Lufttemperatur')
coldest_months = coldest_months.apply(lambda row: f"{int(row['Year'])}/{int(row['Month'])}: {row['Lufttemperatur']:.1f}", axis=1)
print(coldest_months.to_string(index=False))
```

## Q7. Which distribution is suitable to consider to fit the air temperature data?
## How well does the data follow that distribution given some measure, e.g.,
## p-value?

```python
mean_temp = df['Lufttemperatur'].mean()
std_temp = df['Lufttemperatur'].std()
normalized_temp = (df['Lufttemperatur'] - mean_temp) / std_temp
ks_stat, p_value = stats.kstest(normalized_temp, 'norm')
print(f"Kolmogorov-Smirnov statistic: {ks_stat:.4f}")
print(f"p-value: {p_value:.4f}")
```

```python
df = df.dropna()
mean_temp = np.mean(df['Lufttemperatur'])
std_temp = np.std(df['Lufttemperatur'])
ks_result = stats.kstest(df['Lufttemperatur'], 'norm', args=(mean_temp, std_temp))
print(f"Kolmogorov-Smirnov test p-value for normal distribution: {ks_result.pvalue}")
ad_result = stats.anderson(df['Lufttemperatur'], dist='norm')
print(f"Anderson-Darling test statistic for normal distribution: {ad_result.statistic}")
print(f"Critical values for normal distribution: {ad_result.critical_values}")
```

## Q11. Calculate the 95% confidence intervals (CI) for the mean of each daily
## average in Q9 and add it to the line plot that shows the mean values of
## each day of the year 2022. You can, for example, solve this by plotting the
## upper and lower bounds as lines above and below the mean line in your
## plot. Another more nicely looking solution is to visualize the CI as shades
## around the mean line in your plot.

```python
daily_stats_2022 = df_2022.groupby('Datum')['Lufttemperatur'].agg(['mean', 'count', 'std']).reset_index()
daily_stats_2022['CI_lower'] = daily_stats_2022.apply(lambda row: row['mean'] - stats.t.ppf(1 - 0.025, row['count'] - 1) * row['std'] / (row['count'] ** 0.5), axis=1)
daily_stats_2022['CI_upper'] = daily_stats_2022.apply(lambda row: row['mean'] + stats.t.ppf(1 - 0.025, row['count'] - 1) * row['std'] / (row['count'] ** 0.5), axis=1)
plt.figure(figsize=(15, 8))
plt.plot(daily_stats_2022['Datum'], daily_stats_2022['mean'], label='Daily Mean Temperature')
plt.fill_between(daily_stats_2022['Datum'], daily_stats_2022['CI_lower'], daily_stats_2022['CI_upper'], color='gray', alpha=0.3, label='95% CI')
plt.xlabel('Date')
plt.ylabel('Mean Temperature')
plt.title('Daily Mean Temperature in 2022 with 95% Confidence Interval')
plt.legend()
plt.grid()
plt.show()
```
